```python
import pandas as pd
import numpy as np
```

```python
tr_df = pd.read_csv("0000000000002429_training_titanic_x_y_train.csv")
tr_df.describe()
```

```python
te_df= pd.read_csv("0000000000002429_test_titanic_x_test.csv")
te_df.describe()
```

```python
te_df['Survived'] = np.nan
df = pd.concat([tr_df, te_df], axis=0, ignore_index=True)
df.describe()
```

```python
print(df.groupby("Pclass").mean()["Age"])
```

```python
df_y = df.values[:,df.shape[1]-1:df.shape[1]]
del df['Survived']
df_y
```

```python
df
```

```python
df['Age'] = df.groupby('Pclass')['Age'].transform(
    lambda grp: grp.fillna(np.mean(grp))
)
df
df['Sex'].replace(['female','male'],[0,1],inplace=True)
df
```

```python
df.groupby("Embarked").count()
```

```python
df["Embarked"].replace(["S","Q","C"],[0,1,2],inplace=True)
```

```python
df
del df["Cabin"]
del df["Ticket"]
del df["Name"]
```

```python
df.describe()
df["Embarked"].fillna(0,inplace=True)
#S is popular
df.describe()
```

```python
df['Pclass_Pclass'] = df['Pclass'] ** 2
df['Sex_Sex'] = df['Sex'] ** 2
df['Age_Age'] = df['Age'] ** 2
df['SibSp_SibSp'] = df['SibSp'] ** 2
df['Parch_Parch'] = df['Parch'] ** 2
df['Fare_Fare'] = df['Fare'] ** 2
df['Embarked_Embarked'] = df['Embarked'] ** 2
df['Age_Pclass']=df['Age']*df['Pclass']
df
```

```python
from sklearn import preprocessing
```

```python
x_scaled=preprocessing.scale(df.values[:,:])
scaled_df = pd.DataFrame(x_scaled, columns=df.columns[0:df.shape[1]])
scaled_df.describe()
```

```python
df=scaled_df
```

```python
df_ = pd.DataFrame(df_y, columns=tr_df.columns[tr_df.shape[1]-1:tr_df.shape[1]])
df = df.join(df_)
df.head(10)
```

```python
tr_df = df[0:668]
te_df = df[668:891]
del te_df['Survived']
te_df = te_df.reset_index(drop=True)
```

```python
train_data = tr_df
test_data = te_df
```

```python
np.savetxt("new_training_titanic_x_y_train.csv", train_data, delimiter=',')
np.savetxt("new_test_titanic_x_test.csv", test_data, delimiter=',')
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train=train_data.drop("Survived",axis=1)
Y_train=train_data["Survived"]
X_test=test_data
```

```python
Y_train.astype('int')
```

```python
Y_train.describe()
```

```python

logreg = LogisticRegression()
logreg.fit(X_train,Y_train.astype('int'))

Y_pred = logreg.predict(X_test)
logreg.score(X_train,Y_train.astype('int'))
```

```python
np.savetxt("titanic_assignment.csv", Y_pred, delimiter=',')
```
